# Step 4: Handling missing values:

```python
df.isnull().sum()  # check missing values
```

```python
# we'll use median bcoz it is robust to outliers
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))  # we'll replace null value with median of review rating category wise
df.isnull().sum()
```

# Step 5: Transfrom data into proper format:

```python
# change upper case to lower case and remove space by placing under score due to which our data will be consistent and easy to use using python and sql
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df
```

# step 6: feature engineering:

```python
# Create a new column age_group
labels=["Young","Adult","Middle-aged","Senior"]
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)  # qcut splits age into into 4 equal size group based on data distribution or assigns the labels
df[['age','age_group']]
```

```python
# Create a new column purchase_frequency_days
# first convert data of "frequency_of_purchases" into numerical data bcoz it will result in analysis of data much easier
frequency_mapping={   # number of days between purchases
    'Fortnightly' : 14,
    'Weekly':7,
    'Monthly':30,
    'quarterly':90,
    'Bi-Weekly':14,
    'Anually':365,
    'Every 3 Months':90
}

df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
df[['purchase_frequency_days','frequency_of_purchases']]
```

```python
df[['discount_applied','promo_code_used']].head(10)  # both the columns are same so there is no use to take both the table
                                                     # but there might be some case when both the columns are different
                                                     # so let's check it once
```

```python
(df['discount_applied']==df['promo_code_used']).all()   # both the columns are same so we'll remove it
```
